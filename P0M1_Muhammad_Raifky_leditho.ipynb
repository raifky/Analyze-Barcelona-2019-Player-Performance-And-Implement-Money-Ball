## 4. data Cleaning.

The function below is designed to clean the data loaded from the Final_player.csv file. In this process, values containing the character 'm' and the euro currency symbol '€' are removed. After eliminating these elements, the data type is converted from object to float. The market value, which was previously not in a standard numerical format (for example, €180.00), is then converted into a numeric value such as 180000000. This transformation is necessary to ensure that the market value column can be properly calculated and used for descriptive and inferential statistical analysis.



```python
def convert_value(val):
    if isinstance(val, str):
        val = val.replace("€", "").lower().strip()

        if val.endswith("m"):  # juta
            return int(float(val.replace("m", "")) * 1_000_000)

        elif val.endswith("k"):  # ribu
            return int(float(val.replace("k", "")) * 1_000)

    return None  # kalau datanya NaN

df['Market value'] = df['Market value'].apply(convert_value)
df
```

```python
df_filtered_Forward = df[
    (df['Club_x'] == 'Barcelona') & #filtering data dengan pemain yang bermain di barcelona
    (df['Pos'].isin(['FW', 'ST', 'CF', 'WB'])) #filtering posisi depan
][['Player', 'Market value', 'Nation', 'Pos', 'Club_x', 'Leauge', 'Age', 'MP', 'Starts', 'Min', 'Gls', 'Ast', 'Gls90', 'Ast90', 'G+A', 'Gls+Ast', 'Sh', 'SoT%', 'Sh/90', 'SoT/90']] #untuk mengambil beberapa column yang dibutuhkan untuk pemain depan

df_filtered_Forward
```

Based on the descriptive analysis using the IQR method for forward players at FC Barcelona, Lionel Messi appears as a dominant outlier across several key offensive metrics, including assists, assists per 90 minutes, goals plus assists, and shots per 90 minutes. No other forward consistently emerges as an outlier across multiple performance statistics. This indicates that while other players perform within the normal range, Messi’s contribution is significantly higher, resulting in an uneven distribution of offensive performance within the team.

```python
numeric_cols = df_filtered_Forward.select_dtypes(include='number').columns
numeric_cols

def detect_outliers_iqr_df(df_numeric, df_original):
    results = {}

    for col in df_numeric.columns:
        Q1 = df_numeric[col].quantile(0.25)
        Q3 = df_numeric[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # cari outlier pada df_numeric
        mask = (df_numeric[col] < lower_bound) | (df_numeric[col] > upper_bound)

        # ambil baris lengkap dari df_original (supaya Player ikut)
        outliers = df_original[mask]

        print(f"\n===== {col} =====")
        print(f"Q1 = {Q1}")
        print(f"Q3 = {Q3}")
        print(f"IQR = {IQR}")
        print(f"Lower bound = {lower_bound}")
        print(f"Upper bound = {upper_bound}")
        print("Outliers:")

        display(outliers[['Player', col]])

        results[col] = outliers

    return results

# Jalankan fungsi (df numeric + df asli)
iqr_results = detect_outliers_iqr_df(
    df_filtered_Forward[numeric_cols],
    df_filtered_Forward
)

```

This skewness analysis indicate the distribution is right skewed in several offensive metrics, such as what we already give the treshold is Gls90, Ast90, SoT%, and Sh/90. #### Inferential statistic with spearman

After conducting Spearman correlation analysis, a positive correlation between market value and score_forward was observed (ρ = 0.400). However, this relationship is not statistically significant (p = 0.505). Therefore, the null hypothesis cannot be rejected, indicating that there is no sufficient statistical evidence to conclude that market value is associated with attacking performance among Barcelona’s forward players during the observed season.

Hypotesis testing:

H0: There is no significant correlation between market value and score_forward.

H1: There is a significant correlation between market value and score_forward.

So we get H0 for this hypotesis test because p value 0.505 so we receive H0 so the conclusion of this inferential statistic that There is no significant correlation between market value and score_forward that's mean forward player with high market value not always have good stats. 


```python
from scipy.stats import spearmanr

# filter pemain Barcelona posisi forward
df_spearman = df[
    (df['Club_x'] == 'Barcelona') &
    (df['Pos'].isin(['FW', 'ST', 'CF', 'WB']))
][['Market value', 'Score_forward']].dropna()

rho, p_value = spearmanr(
    df_spearman['Market value'],
    df_spearman['Score_forward']
)

print(f"Spearman correlation (ρ): {rho:.3f}")
print(f"P-value: {p_value:.5f}")
```

```python
numeric_cols_def = df_filtered_Defender.select_dtypes(include='number').columns
numeric_cols_def

def detect_outliers_iqr_df(df_numeric, df_original):
    results = {}

    for col in df_numeric.columns:
        Q1 = df_numeric[col].quantile(0.25)
        Q3 = df_numeric[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Mask outlier
        mask = (df_numeric[col] < lower_bound) | (df_numeric[col] > upper_bound)

        # Ambil baris asli dengan kolom Player
        outliers = df_original[mask]

        print(f"\n===== {col} =====")
        print(f"Q1 = {Q1}")
        print(f"Q3 = {Q3}")
        print(f"IQR = {IQR}")
        print(f"Lower bound = {lower_bound}")
        print(f"Upper bound = {upper_bound}")
        print("Outliers:")
        display(outliers[['Player', col]])

        results[col] = outliers

    return results
iqr_results_def = detect_outliers_iqr_df(
    df_filtered_Defender[numeric_cols_def],
    df_filtered_Defender
)
```

#### Inferential statistic with spearman

The Spearman correlation analysis between market value and Score_defender yields a coefficient of −0.316 with a p-value of 0.684, indicating a weak negative and statistically insignificant relationship. This result suggests that higher market value does not correspond to higher defensive performance scores, highlighting the importance of tactical role and collective defensive structure over individual valuation.

Hypotesis testing:

H0: There is no significant correlation between market value and score_defender.

H1: There is a significant correlation between market value and score_defender.

So we get H0 for this hypotesis test because p value 0.505 so we receive H0 so the conclusion of this inferential statistic that There is no significant correlation between market value and score_forward that's mean forward player with high market value not always have good stats. 

```python
df_spearman = df[
    (df['Club_x'] == 'Barcelona') &
    (df['Pos'].isin(['DF', 'DF,MF']))
][['Market value', 'Score_defender']].dropna()

rho, p_value = spearmanr(
    df_spearman['Market value'],
    df_spearman['Score_defender']
)

print(f"Spearman correlation (ρ): {rho:.3f}")
print(f"P-value: {p_value:.5f}")
```
